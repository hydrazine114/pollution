```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0.02, 1e1, 100000)

df = pd.DataFrame({'x': x, '-1/x': -1/x, '-1/x^2': -1/x**2, '-1/x^3': -1/x**3, '-1/x^6': -1/x**6, 'exp': 1e-1*np.exp(1/x)})
df['s'] = df.iloc[:, 1] + df.iloc[:, -1]
a = sns.color_palette("magma", 9)
fig, ax = plt.subplots(dpi=200)
ax.plot(df.x, df['-1/x'], label=r'$V_{ii} \sim \frac{1}{r}$', color=a[2])
ax.plot(df.x, df['exp'], label=r'$V_{id} \sim \frac{1}{r^2}$', color=a[4])
ax.plot(df.x, df.iloc[:, 1] + df.iloc[:, -1], label=r'$V_{dd} \sim \frac{1}{r^3}$', color=a[5])
ax.set_xlim([-0.1, 10])
ax.set_ylim([-5, 5])

ax.hlines(-0.0, 0, 1e5, color='k', linewidth=0.7)
ax.vlines(0, -1e5, 1e5, color='k', linewidth=0.7)

# fig.legend(frameon=False, bbox_to_anchor=[0.8, 0.4])

ax.spines['bottom'].set_color('white')
ax.spines['right'].set_color('white')
ax.spines['top'].set_color('white')
ax.spines['left'].set_color('white')


ax.set_xticks([])
ax.set_yticks([])
ax.text(-0.5, 5, r'$E$')
ax.text(9.8, 0.8, r'$r$')
```
